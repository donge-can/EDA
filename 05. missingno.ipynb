```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn')
sns.set(font_scale=2.5)

import missingno as msno

#ignore warnings
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
path = 'C:\\Users\\user\\TIL\\TIL\\kaggle\\01. titanic\\'
df_train = pd.read_csv(path+'train.csv')
df_test = pd.read_csv(path+'test.csv')
```

<br>

## 결측치 시각화
<br>

### `Missingno` 모듈
<br>

### 방법 01 - msno.matrix

```python
msno.matrix(df=df_train.iloc[:, :], figsize =(8,8), color = (0.8, 0.5, 0.2))
```

<br>

### 방법 02 - msno.bar

```python
msno.bar(df_train.sample(500), log = True)

# msno.bar(df_train.sample(500))

# msno.bar(df_train) 

# df_train.sample(500) : 무작위로 500개 추출
```

```python
msno.bar(df=df_test, figsize = (8,8), color = (0.8, 0.5, 0.2))
```

<br>

### 방법 03 - msno.heatmap

```python
msno.heatmap(df_train)

# 결측치가 하나도 없거나 아예 비어 있는 변수들은 의미있는 상관관계를 나타내기 힘들기 때문에 자동으로 제거됨
# 변수들 사이의 데이터 상관 관계를 파악해서 특정 변수를 선택하는 데 효과적이긴 하나,
# 더 큰 관계를 나타내기 위한 목적으로는 설명력이 부족하다. 또한, 방대한 데이터셋을 지원하지 않는다는 단점이 있음
```

<br>

### 방법 04- msno.dendrogram
